```python
#This is a script to solve for the motion of a massive rod double pendulum

import numpy as np
from scipy.integrate import odeint
import MBstructs as MB
import matplotlib.pyplot as plt
import MultiBodyFuncts as MBF
import pycuda.driver as cuda
import pycuda.autoinit
from pycuda.compiler import SourceModule
%matplotlib inline
```

```python
#System definitions and initial conditions
g=9.81
q1init= np.pi/4
q2init= np.pi/2
timesteps=1
Time=np.linspace(0,10,timesteps)
xinit=np.array(( q1init, q2init-q1init,0,0))
B1=MB.Body()
B2=MB.Body()
B1.m=5.0
B2.m=1.0
B1.l=5.0
B2.l=1.0
B1.Inertia(B1.m*(B1.l**2)/12,0,B1.m*(B1.l**2)/12)
B2.Inertia(B2.m*(B2.l**2)/12,0,B2.m*(B2.l**2)/12)
B1.bvectors(np.array((0,-B1.l,0)),np.array((0,-B1.l/2,0)))
B2.bvectors(np.array((0,-B2.l,0)),np.array((0,-B2.l/2,0)))
```

```python
#Figure depicting the system
la1x=la1y=0
la2x=B1.l*np.sin(q1init)
la2y=-B1.l*np.cos(q1init)
lb2x=la2x+B2.l*np.sin(q2init)
lb2y=la2y-B2.l*np.cos(q2init)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.plot([la1x,la2x],[la1y,la2y],[la2x,lb2x],[la2y,lb2y],'b',[-4,7],[0,0])

plt.axis([-1,6,-6.5,.5])
plt.show()
```

```python
#Function to be integrated

def funct(xinits,t):
    print("hi")
    
    #Coordinates
    x=np.array((xinits[0],xinits[1]))
    xdot=np.array((xinits[2],xinits[3]))
    
    #Generalized Coordinates
    xgen=np.array((0,0,x[0],0,0,x[1]))
    xgendot=np.array((0,0,xdot[0],0,0,xdot[1]))
    num=np.array((2,5))
  
    #Number of bodies and number of degrees of freedom
    N=2
    n=2
   
    #List of bodies used to loop through the system
    sys=[B1,B2]
            
    #Direction Cosine Matrices
    B1.C0=MBF.simdcm(x[0],np.array((0,0,1)))
    B2.C0=MBF.simdcm(x[1]+x[0],np.array((0,0,1)))
    
    #Inertia dyadic matrices
    for i in range(0,N):
        sys[i].Inertia2()       
    #Partial Angular Velocities    
    for i in range(0,N):
        j=1
        ws=np.zeros((3*N,3))
        ws[:3,:]=np.identity(3)
        if i>0:
            while j < N:
                ws[3*j:3*j+3,:]=sys[j-1].C0
                j=j+1
        sys[i].Pw=ws
               
    #Angluar Velocities and Derivatives of DCMs
    for i in range(0,N):
        sys[i].angvel(xgendot)
        sys[i].Sderiv()
               
    #Time Derivative of Partial Angular Velocities
    for i in range(0,N):
        j=1
        wdots=np.zeros((3*N,3))
        wdots[:3,:]=np.zeros((3,3))
        if i>0:
            while j < N:
                wdots[3*j:3*j+3,:]=sys[j-1].C0dot
                j=j+1
        sys[i].Pwdot=wdots
        
    #Body Vector skew Matrices Derivatives
    B1.Srdot=np.zeros((3,3))
    B1.Sqdot=B1.Srdot
    B2.Srdot=B1.Srdot
          
    #Partial Velocity Matrices
    for i in range(0,N):
        j=0
        Vs=np.zeros((3*N,3))
        while j < i:
            Vs[j*3:j*3+3,:]=np.dot(sys[j].Sq,sys[j].C0)
            j=j+1
        Vs[j*3:j*3+3,:]=np.dot(sys[i].Sr,sys[i].C0)
        sys[i].PV=Vs
               
    #Derivative of Partial Velocity Matrices
    for i in range(0,N):
        j=0
        Vdots=np.zeros((3*N,3))
        while j < i:
            Vdots[j*3:j*3+3,:]=np.dot(sys[j].Sq,sys[j].C0dot)+np.dot(sys[j].Sqdot,sys[j].C0)
            j=j+1
        Vdots[j*3:j*3+3,:]=np.dot(sys[i].Sr,sys[i].C0dot)+np.dot(sys[i].Srdot,sys[i].C0)
        sys[i].PVdot=Vdots
        
   
    # w transformation matrix
    w=np.zeros((3*N,3*N))
    for i in range(0,N):
        j=i*3
        while j<3*N:
            w[3*i:3*i+3,j:j+3]=sys[i].C0
            j=j+3
         
    #y coordinates
    y=np.dot(xgendot,w)
    
    B1.force=np.array((0,-1*B1.m*9.81,0))
    B2.force=np.array((0,-1*B2.m*9.81,0))
    
         
    #Equations of Motion
    a=b=c=d=f=0  
    for i in range(0,n):
        
        #First coefficient
        a=a+sys[i].m*np.dot(np.dot(w,sys[i].PV),np.transpose(np.dot(w,sys[i].PV)))
        a=a+np.dot(sys[i].Pw,np.dot(sys[i].I0T,np.transpose(sys[i].Pw)))
        
        #Second Coefficient
        b=b+sys[i].m*np.dot(np.dot(w,sys[i].PV),np.transpose(np.dot(w,sys[i].PVdot)))
        b=b+np.dot(sys[i].Pw,np.dot(sys[i].I0T,np.transpose(sys[i].Pwdot)))
        
        #Third Coefficient
        c=c+np.dot(sys[i].Pw,np.dot(sys[i].I0T,np.dot(np.transpose(sys[i].W0),np.transpose(sys[i].Pw))))
        
        #Force
        f=f+np.dot(np.dot(w,sys[i].PV),sys[i].force)
    
    #Combine c and b
    d=b+c
                
    #M is the matrix used to create equations to integrate            
    M=np.zeros((n,2*n+1))
    #num = [i for i in range(0,n*3) if xgen[i] != 0]
    for i in range(0,n):
        for j in range(0,n):
            M[i,j]=a[num[i],num[j]]
            M[i,j+n]=d[num[i],num[j]]
        M[i,2*n]=f[num[i]]
        
    #Put the system in a form suitable for integration
    for i in range(0,n):
        M[i,:]=M[i,:]/M[i,i]
        for j in range(0,n):
            if i !=j:
                M[j,:]=M[j,:]-(M[i,:]*M[j,i])
                
    d_dt=np.zeros((2*n))

    for i in range(0,n):
        d_dt[i]=xgendot[num[i]]
        d_dt[i+n]=M[i,2*n]
    
    for i in range(0,n):
        for j in range(0,n):
            d_dt[i+n]=d_dt[i+n]-xgendot[num[j]]*M[i,j+n] 
    return d_dt
    
```

```python

#Integrate function and plot results

yy=odeint(funct,xinit,Time)
theta1=yy[:,0]
theta2=yy[:,1]+theta1

plt.plot(Time,theta1,Time,theta2)
plt.xlabel("Time [s]")
plt.ylabel("Angle [Rad]")
plt.title("System Response")
plt.legend(("Theta1","Theta2"),"best")

plt.show()
```
