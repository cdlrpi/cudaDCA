```python
#System definitions and initial conditions
g=9.81
q1init= np.pi/4
q2init= np.pi/2
timesteps=5
Time=np.linspace(0,10,timesteps)
initialconditions=np.array(( q1init, q2init,0,0))

#Bodies and Joints
B1=MB.Body()
B2=MB.Body()
J0=MB.Joint()
J1=MB.Joint()
J2=MB.Joint()

#P and D matrices
J0.PDmatrix(np.array((0,0,1,0,0,0)),1)
J1.PDmatrix(np.array((0,0,1,0,0,0)),1)
J2.PDmatrix(np.array((0,0,1,1,1,0)),3)#unsure of P for the last joint
print(J2.P)
B1.m=5.0
B2.m=1.0
B1.l=5.0
B2.l=1.0

#apply the active force
B1.gravity(np.array((0,-1,0)))
B2.gravity(np.array((0,-1,0)))
#constant inertia properties
B1.Inertia(B1.m*(B1.l**2)/12,1,B1.m*(B1.l**2)/12)
B2.Inertia(B2.m*(B2.l**2)/12,1,B2.m*(B2.l**2)/12)

#these are used for system plot
B1.bvectors(np.array((0,-B1.l,0)),np.array((0,-B1.l/2,0)))
B2.bvectors(np.array((0,-B2.l,0)),np.array((0,-B2.l/2,0)))

#put the system into two lists
bs=[B1,B2]
js=[J0,J1,J2]
```

```python
#Recursive DCA Algorithm (part 1?)
  
#Right now BC=1 means free, =2 means pin, will be more robust
def recursiveDCA(n,i,bodies,joints,BC1,BC2,xinits): 
              
    if len(bodies) !=1: #if there are still more than one bodies in the system
        j=0
        odd = 0
        #check if there are an odd number of bodies
        #create a new list of bodies 
        #newbds is the list that will contain the newly assembled bodies
        if (len(bodies)%2)==0:
            newbds=[MB.Body]*np.int32(len(bodies)/2)
        else:
            odd=1
            newbds=[MB.Body]*np.int32(math.trunc(len(bodies)/2)+1)
        #loop through all of the new bodies
        while j < len(newbds):
            #if there are an odd # of bodies and we're at the last 
            #body in newbds, add the last body to newbds
            if j == len(newbds)-1 and odd == 1:
                newbds[j].z11=bodies[2*j].z11
                newbds[j].z12=bodies[2*j].z12
                newbds[j].z21=bodies[2*j].z21
                newbds[j].z22=bodies[2*j].z22
                newbds[j].z13=bodies[2*j].z13
                newbds[j].z23=bodies[2*j].z23
                newbds[j].m=bodies[2*j].m
            else:#otherwise, calculate the new zetas for the newly assembled bodies
                newbds[j].m=bodies[2*j].m+bodies[2*j+1].m         
                #X, Q, and Y  are  intermediate quantities
                newbds[j].X=np.dot(np.transpose(joints[2*j+1].D),np.dot(bodies[2*j+1].z11+bodies[2*j].z22,joints[2*j+1].D))
                newbds[j].Xinv=np.linalg.inv(newbds[j].X)
                newbds[j].W=np.dot(joints[2*j+1].D,np.dot(newbds[j].Xinv,np.transpose(joints[2*j+1].D)))
                newbds[j].Y=np.dot(newbds[j].W,bodies[2*j].z23-bodies[2*j+1].z13)#ommitted pdot*u
                #assemble the bodies based on the formulas
                newbds[j].z11=bodies[2*j].z11-np.dot(bodies[2*j].z12,np.dot(newbds[j].W,bodies[2*j].z21))
                newbds[j].z12=np.dot(bodies[2*j].z12,np.dot(newbds[j].W,bodies[2*j+1].z12))
                newbds[j].z21=np.dot(bodies[2*j+1].z21,np.dot(newbds[j].W,bodies[2*j].z21))
                newbds[j].z22=bodies[2*j+1].z22-np.dot(bodies[2*j+1].z21,np.dot(newbds[j].W,bodies[2*j+1].z12))
                newbds[j].z13=bodies[2*j].z13-np.dot(bodies[2*j].z12,newbds[j].Y)
                newbds[j].z23=bodies[2*j+1].z23+np.dot(bodies[2*j+1].z21,newbds[j].Y)
            j=j+1
        #now find the joints corresponding to the new bodies
        newjs=[MB.Joint]*(len(newbds)+1)
        #loop that brings along the P D and u matrices from previous joints
        for j in range(0,len(newjs)):
            if j==len(newjs)-1 and odd ==1:
                newjs[j].D=joints[len(joints)-1].D
                newjs[j].P=joints[len(joints)-1].P
                newjs[j].uj=joints[len(joints)-1].uj
            else:
                newjs[j].D=joints[2*j].D
                newjs[j].P=joints[2*j].P
                newjs[j].uj=joints[2*j].uj
                
                
        #This is the recursive call.  This will return a list of the form:
        #[A1,F1c,A2,F2c ...] where the A's and Fc's correspond to the 
        #accellerations and constraint forces in each joint
        #The function is given the original number of bodies, 
        #the list of new bodies, the list of new joints, the boundary conditions,
        #and the state variables at that timestep
        #At this point this function will repeat itself in a loop until there is
        #one body left

        sol=recursiveDCA(n,i+1,newbds,newjs,BC1,BC2,xinits)
       
        
        #Now that these constraint forces and accellerations are found, 
        #we can solve for these values for the larger number of bodies
        newsol=[];
        newsol.append(sol[0])
        newsol.append(sol[1])
        
       
        
        #In this loop I start with the first body and find the force and 
        #acceleration on its other handle. These values are added to the new
        #solution list, along with the next values in sol.
        #I'm not sure if I'm doing this right because I never use the last Fc and A
        #in sol.  Should I be working through the list backwards too to get the
        #"other side" of each joint?
        flag=0
        for j in range(0,len(newbds)):
            if ~(odd==1 and j==len(newbds)-1):
                k=len(newsol)-2
                #F=np.dot(np.linalg.inv(bodies[2*j].z12),newsol[k]-bodies[2*j].z13-np.dot(bodies[2*j].z11,newsol[k+1]))
                F=-1*(np.dot(joints[j+1].D,np.dot(newbds[j].Xinv,np.dot(np.transpose(joints[j+1].D),(np.dot(bodies[2*j].z21,newsol[k+1])-np.dot(bodies[2*j+1].z12,sol[2*(j+1)+1])+bodies[2*j].z23-bodies[2*j+1].z13)))))
                A2=np.dot(bodies[2*j].z21,newsol[k+1])+np.dot(bodies[2*j].z22,F)+bodies[2*j].z23
               
                A1=np.dot(bodies[2*j+1].z11,-1*F)+np.dot(bodies[2*j+1].z12,sol[4*j+3])+bodies[2*j+1].z13
 
                newsol.append(A2)
                newsol.append(F)
                newsol.append(A1)
                newsol.append(-1*F)
                newsol.append(sol[2*(j+1)])
                newsol.append(sol[2*(j+1)+1])
            elif odd==1 and j==len(newbds)-2:
                newsol.append(sol[2*(j+1)])
                newsol.append(sol[2*(j+1)+1])
                flag=1
                
            if odd ==1 and j!=len(newbds)-1  and flag !=1:   
                newsol.append(sol[2*(j+1)+2])
                newsol.append(sol[2*(j+1)+3])
        
        if i ==0:
            for j in range (1,n+3):
                del newsol[j]
                  
            return newsol
        else:
        #Finally return all of the forces and accellerations of each joint
        #at this level of recursion.  If this is the first level of recursion
        #this will be the final return
            return newsol
    
    #This is where it goes when theres only 1 body left
    #These are the only boundary conditions ive made so far
    elif BC1==2 and BC2==1:
        #Fc2 is zero because that end is free
        Fc2=np.zeros((6))
        #Because the "first" end is pinned, it will
        #have no translational motion, and support no 
        #moments, so mat is used to solve for the three translational
        #forces of Fc1
        mat=np.zeros((3,4))
        mat[:,:3]=bodies[0].z11[3:,3:]
        mat[:,3]=-1*bodies[0].z13[3:]
        
        #cool little loop that takes a matrix
        #and solves the system of equations
        print(mat)
        for s in range(0,3):
            mat[s,:]=mat[s,:]/mat[s,s]
            for j in range(0,3):
                if s !=j:
                    mat[j,:]=mat[j,:]-(mat[s,:]*mat[j,s])
        
        print(mat)
        Fc1=np.zeros_like(Fc2)
        Fc1[3:]=mat[:,3]
        
        #solve for the A's given Fc1 and Fc2=0
        A1=np.dot(bodies[0].z11,Fc1)+bodies[0].z13
        A2=np.dot(bodies[0].z21,Fc1)+bodies[0].z23
        sol=[A1,Fc1,A2,Fc2]
        
        #return the forces and accellerations at the end joints
        #and begin disassembly
        return sol
        
```

```python
def initialize(b,x):
    
    j=0
    while j<n:#loop through all the bodies
        b[j].C0=MBF.simdcm(x[j],np.array((0,0,1)))#DCM
        b[j].rs(b[j].l/2.0,np.array((0,-1,0)))#position vectors
        b[j].Inertia2()#transformed inertia matrix
        b[j].Mmatrix()#Mass matrix and its inverse <-- problems here
        b[j].shifters()#S matrices
        b[j].zs()#zeta values
        j=j+1
```

```python
#Function to be integrated
#This function calls recursiveDCA, and finds the u matrix for each joint
#not sure if the u's are constructed this way
def funct(state,t,n,i,bbs,jjs,BC1,BC2):
    m=0;
    p=0;
    l=n;
    for k in Dofperj:
        jjs[m].uj=np.zeros((k))
        p=0
        while p<k:
            jjs[m].uj[p]=state[l]
            p=p+1
            l=l+1
        m=m+1
    initialize(bbs,state)
    ll=recursiveDCA(n,0,bbs,jjs,BC1,BC2,state)
    
    #right now I'm taking the udots as the angular acceleration about the n3
    #axis for the first and last joints.  This will be changed to work with 
    #length n systems but im not sure if exactly true
    d_dt=np.zeros((2*n))
    for j in range(0,n):
        if j == 0:
            A1=ll.pop(0)
            d_dt[j+n]=np.dot(np.transpose(jjs[j].P),A1)
        else:
            A2= ll.pop(0)
            A1=ll.pop(0)
            d_dt[j+n]=np.dot(np.transpose(jjs[j].P),(A1-A2))
    d_dt[:n]=state[n:]
   
  
    return d_dt
    
```

```python
#Integrate function and plot results
n=2
i=0
BC1=2
BC2=1
yy=odeint(funct,initialconditions,Time,(n,i,bs,js,BC1,BC2))
theta1=yy[:,0]
theta2=yy[:,1]

plt.plot(Time,theta1,Time,theta2)
plt.xlabel("Time [s]")
plt.ylabel("Angle [Rad]")
plt.title("System Response")
plt.legend(("Theta1","Theta2"),"best")

plt.show()
print("The above graph is wrong because the mass matrix is wrong to allow inversion")
```
