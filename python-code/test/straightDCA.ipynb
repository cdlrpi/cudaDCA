```python
# Straight DCA
#---------------
# This is a script to solve for the motion of a 
# massive rod double pendulum using DCA
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt

import MBstructs as MB
import MultiBodyFuncts as MBF
import RK45 

%matplotlib inline
```

```python
# System Definition
#--------------------

#number of bodies
n=2

#Bodies and Joints creation
B1=MB.Body()
B2=MB.Body()
J0=MB.Joint()
J1=MB.Joint()

#Joint Properties
J0.PDmatrix(np.array((0,0,1,0,0,0)),1)
J1.PDmatrix(np.array((0,0,1,0,0,0)),1)

#Body Properties
B1.m=1.0
B2.m=1.0
B1.l=1.0
B2.l=1.0
B1.Inertia(B1.m*(B1.l**2)/12,1,B1.m*(B1.l**2)/12)
B2.Inertia(B2.m*(B2.l**2)/12,1,B2.m*(B2.l**2)/12)

#Create a list of Bodies and Joints
bs=[B1,B2]
js=[J0,J1]
```

```python
# Initial Conditions
#--------------------

q1init= np.pi/2
q2init= 0
q1dotinit=0
q2dotinit=0
initialconditions=np.array(( q1init, q2init,q1dotinit,q2dotinit))

#Length of time of the simulation

Time=np.arange(0,5,.001)
```

```python
# System Depiction
#------------------
la1x=la1y=0
la2x=B1.l*np.sin(q1init)
la2y=-B1.l*np.cos(q1init)
lb2x=la2x+B2.l*np.sin(q2init)
lb2y=la2y-B2.l*np.cos(q2init)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.plot([la1x,la2x],[la1y,la2y],[la2x,lb2x],[la2y,lb2y],'b',[-4,7],[0,0])
plt.axis([-1,6,-6.5,.5])
plt.show()
```

```python
# Straight DCA Algorithm
#------------------------
#This function applies the DCA Algorithm to a compound double 
#pendulum system without the use of loops or recursion. 

def DCA(bodies,joints,inits):
    
    #Create the new body
    newbds=MB.Body()

    #X, Q, and Y  are  intermediate quantities
    newbds.X=np.dot(np.transpose(joints[1].D),np.dot(bodies[1].z11+bodies[0].z22,joints[1].D))
    newbds.Xinv=np.linalg.inv(newbds.X)
    newbds.W=np.dot(joints[1].D,np.dot(newbds.Xinv,np.transpose(joints[1].D)))
    newbds.Y=np.dot(newbds.W,bodies[0].z23-bodies[1].z13)#ommitted pdot*u
    
    #assemble the bodies based on the formulae
    newbds.z11=bodies[0].z11-np.dot(bodies[0].z12,np.dot(newbds.W,bodies[0].z21))
    newbds.z12=np.dot(bodies[0].z12,np.dot(newbds.W,bodies[1].z12))
    newbds.z21=np.dot(bodies[1].z21,np.dot(newbds.W,bodies[0].z21))
    newbds.z22=bodies[1].z22-np.dot(bodies[1].z21,np.dot(newbds.W,bodies[1].z12))
    newbds.z13=bodies[0].z13-np.dot(bodies[0].z12,newbds.Y)
    newbds.z23=bodies[1].z23+np.dot(bodies[1].z21,newbds.Y)

    #Constraint force on the free end is zero
    Fc2=np.zeros((6))
    
    #Because the first joint is a pin, it will
    #have no translational motion, and support no 
    #moments, so mat is used to solve for the two translational
    #forces of Fc1
    mat=np.zeros((3,4))
    mat[:,:3]=newbds.z11[3:,3:]
    mat[:,3]=-1*newbds.z13[3:]
    
    #Loop that puts the matrix in reduced row echelon form
    for s in range(0,3):
        mat[s,:]=mat[s,:]/mat[s,s]
        for j in range(0,3):
            if s !=j:
                mat[j,:]=mat[j,:]-(mat[s,:]*mat[j,s])
    
    
    Fc1=np.zeros_like(Fc2)
    Fc1[3:]=mat[:,3]
    
    #Solve for the acceleration at each end given Fc2=0
    A1=np.dot(newbds.z11,Fc1)+newbds.z13
    A2=np.dot(newbds.z21,Fc1)+newbds.z23
    
    #Find the force in the joint connecting the rods
    F=-1*(np.dot(joints[1].D,np.dot(newbds.Xinv,np.dot(np.transpose(joints[1].D),(np.dot(bodies[0].z21,Fc1)-np.dot(bodies[1].z12,Fc2)+bodies[0].z23-bodies[1].z13)))))
    
    #Find the acceleration at each side of the connecting joint
    A12=np.dot(bodies[0].z21,Fc1)+np.dot(bodies[0].z22,F)+bodies[0].z23
    A21=np.dot(bodies[1].z11,-1*F)+np.dot(bodies[1].z12,Fc2)+bodies[1].z13
    
    #Use the accelerations to find the udots
    u1dot=np.dot(np.transpose(joints[0].P),A1)
    u2dot=np.dot(np.transpose(joints[1].P),(A21-A12))
    
    #Assemble and return the vector of state values
    d_dt=np.zeros((4))
    d_dt[:2]=inits[2:]
    d_dt[3]=u2dot
    d_dt[2]=u1dot
    return d_dt
```

```python
# Initialization Function
#-------------------------
#This function prepares the system for the DCA algorithm 

def initialize(b,x):
    
    #Direction Cosine Matrices
    b[0].C0=MBF.simdcm(x[0],np.array((0,0,1)))
    b[1].C0=MBF.simdcm(x[1]+x[0],np.array((0,0,1)))
    
    #Angular Velocity with respect to the newtonian frame
    b[0].w=np.array((0,0,x[2]))
    b[1].w=np.array((0,0,x[3]+x[2]))
    
    #Position Vectors pointing from joints to centers of mass
    b[0].rs(b[0].l/2.0,np.array((0,-1,0)))
    b[1].rs(b[1].l/2.0,np.array((0,-1,0)))
   
    j=0
    while j<2:#Loop through all the bodies 
        b[j].Inertia2()#Transform the Inertia Matrix 
        b[j].Mmatrix()#Create the Mass matrix and its inverse
        b[j].Forces(np.array((0,-1,0)))#Apply the state determined forces
        b[j].shifters()#Find shifter matrices
        b[j].zs()#Calculate initial zeta values
        j=j+1

   
```

```python
# Helper Function
#-------------------------
#This function provides a buffer between the dca algorithm and the 
#integrator call.  
def funct(state,t,n,bbs,jjs):
   
    #Initialize the system for this timestep
    initialize(bbs,state)

    #Perform DCA
    d_dt=DCA(bbs,jjs,state)

    return d_dt  
```

```python
# Integration
#-------------

#odeint is the numerical integrator used
yy=odeint(funct,initialconditions,Time,(n,bs,js))
```

```python
# Energy Calculation
#--------------------
#The energy of the system is calculated and plotted

energy=MBF.PendEnergy(yy,bs)
KE=energy[:,0]
PE=energy[:,1]
TE=energy[:,2]

plt.plot(Time,TE-TE[0])
plt.xlabel("Time [s]")
plt.ylabel("energy")
plt.title("System Energy")
plt.show()

plt.plot(Time,PE,Time,KE)
plt.xlabel("Time[s]")
plt.ylabel("energy")
plt.title("Kinetic and Potential Energy")
plt.show
```

```python
# Solution Plot
#--------------------

plt.plot(Time,yy[:,:n])
plt.xlabel("Time [s]")
plt.ylabel("Angle [Rad]")
plt.title("System Response")
plt.legend(("Theta1","Theta2"),"best")

plt.show()

plt.plot(Time,yy[:,n:])
plt.xlabel(("Time[s]"))
plt.ylabel(("Generalized Speeds [Rad/s]"))
plt.title("System Response")

plt.show()
```
