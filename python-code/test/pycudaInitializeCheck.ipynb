```python
q1init= np.pi/2
q2init=0
q1dotinit=0
q2dotinit=0
initialconditions=np.array(( q1init, q2init,q1dotinit,q2dotinit))

#number of bodies
n=2

#Bodies and Joints creation
B1=MB.Body()
B2=MB.Body()
B3=MB.Body()
B4=MB.Body()

B1.m=1.0
B2.m=1.0
B3.m=1.0
B4.m=1.0
B1.l=1.0
B2.l=1.0
B3.l=1.0
B4.l=1.0
B1.Inertia(B1.m*(B1.l**2)/12,1,B1.m*(B1.l**2)/12)
B2.Inertia(B2.m*(B2.l**2)/12,1,B2.m*(B2.l**2)/12)
B3.Inertia(B3.m*(B3.l**2)/12,1,B3.m*(B3.l**2)/12)
B4.Inertia(B4.m*(B4.l**2)/12,1,B4.m*(B4.l**2)/12)
bs=[B1,B2]
bbs=[B3,B4]
```

```python
# Initialization Function
#-------------------------
#This function prepares the system for the DCA algorithm 

def pycudaInitialize(b,x,n):
    
    x_gpu = x.astype(np.float32)
    m_gpu = np.zeros((n))
    l_gpu = np.zeros((n))
    I_gpu = np.zeros((3,3*n))
    for i in range (0,n):
        m_gpu[i]=b[i].m
        l_gpu[i]=b[i].l
        I_gpu[:,3*i:3*i+3]=b[i].I
    m_gpu=m_gpu.astype(np.float32)
    l_gpu=l_gpu.astype(np.float32)
    I_gpu=I_gpu.astype(np.float32)
    gpu_out=np.zeros((6,26*n)).astype(np.float32)
    Initialize(cuda.In(x_gpu),cuda.In(m_gpu), cuda.In(l_gpu),cuda.In(I_gpu),cuda.Out(gpu_out),block=(6,6,1),grid=(n,1,1))

    k=0
    index1 = 0
    while k<n:
        b[k].z11=gpu_out[:,index1:index1+6]
        index1 +=6
        b[k].z12=gpu_out[:,index1:index1+6]
        index1 +=6
        b[k].z13=gpu_out[:,index1]
        index1+=1
        b[k].z21=gpu_out[:,index1:index1+6]
        index1+=6
        b[k].z22=gpu_out[:,index1:index1+6]
        index1+=6
        b[k].z23=gpu_out[:,index1]
        index1+=1
        print(b[k].z11)
        k+=1
```

```python
def init(b,x,n):   
    val1=0
    val2=np.zeros((3))
    val3=np.zeros((3))
    for k in range(0,n):
        val1=val1+x[k]
        b[k].C0=MBF.simdcm(val1,np.array((0,0,1)))
        val2=val2+np.array((0,0,x[k+n]))
        b[k].w=val2
        b[k].w1=np.array((0,0,x[k+n]))
        b[k].rs(b[k].l/2.0,np.array((0,-1,0)))
        b[k].Vcm=val3+np.cross(b[k].w,b[k].r10)
        val3=val3+np.cross(b[k].w,b[k].r12)
        b[k].Inertia2()#transformed inertia matrix
        b[k].Mmatrix()#Mass matrix and its inverse
        b[k].Forces(np.array((0,-1,0)))
        b[k].shifters()#S matrices
        b[k].zs()#zeta values
 
    
    
```
