```python
#This is a script to solve for the motion of a simple double pendulum

import numpy as np
from scipy.integrate import odeint
import MBstructs as MB
import matplotlib.pyplot as plt
import MultiBodyFuncts as MBF
import pycuda.driver as cuda
import pycuda.autoinit
from pycuda.compiler import SourceModule
%matplotlib inline
```

```python
#System definitions and initial conditions
la=5.0
lb=1.0
g=9.81
q1init= np.pi
q2init= np.pi/2.0
timesteps=1000
Time=np.linspace(0,10,timesteps)
m1=5
m2=1
g=9.81
xinit=np.array(( q1init, q2init-q1init,0,0))
```

```python
#Figure depicting the system
la1x=la1y=0
la2x=la*np.sin(q1init)
la2y=-la*np.cos(q1init)
lb2x=la2x+lb*np.sin(q2init)
lb2y=la2y-lb*np.cos(q2init)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.plot([la1x,la2x],[la1y,la2y],[la2x,lb2x],[la2y,lb2y],'b',[-4,7],[0,0])
circ1=plt.Circle((la2x,la2y),.2,color='b')
circ2=plt.Circle((lb2x,lb2y),.2,color='b')
ax.add_patch(circ1)
ax.add_patch(circ2)
plt.axis([-1,6,-6.5,.5])
plt.show()
```

```python
#Function to be integrated

def funct(xinits,t):
    x=np.array((xinits[0],xinits[1]))
    xdot=np.array((xinits[2],xinits[3]))
    m1=5
    m2=1
    #Number of bodies and number of degrees of freedom
    N=2
    n=2
            
        #Direction Cosine Matrices
    San=MBF.simdcm(x[0],np.array((0,0,1)))
    Sbn=MBF.simdcm(x[1]+x[0],np.array((0,0,1)))
    
            
        #Generalized Coordinates
    xgen=np.array((0,0,x[0],0,0,x[1]))
    xgendot=np.array((0,0,xdot[0],0,0,xdot[1]))
         
        #Partial Angular Velocities
    w1=np.zeros((3*N,3))
    w2=np.zeros_like(w1)
          
    w1[:3,:]=np.identity(3)
    w2[:3,:]=np.identity(3)
    w2[3:6,:]=San
    
         
        #Angluar Velocities
    w1n=np.dot(xgendot,w1)
    w2n=np.dot(xgendot,w2)
    
          
        #Scew Matrices of Angluar Velocites
    W01=MBF.skew(w1n)
    W02=MBF.skew(w2n)
    
    
           
        #Derivatives of DCMs
    Sandot=np.dot(np.transpose(W01),San)
    Sbndot=np.dot(np.transpose(W02),Sbn)
    
         
           
        #Time Derivative of Partial Angular Velocities
    w1dot=np.zeros_like(w1)
    w2dot=np.zeros_like(w2)
    w2dot[3:,:]=Sandot
          
        #Body Vectors
    q1=np.array((0,0,0))
    q2=np.array((0,-la,0))
    q3=np.array((0,-lb,0))
    r1=q2
    r2=q3
            
        #Body Vector skew Matrices
    Sq2=MBF.skew(q2)
    Sq3=MBF.skew(q3)
    Sr1=MBF.skew(r1)
    Sr2=MBF.skew(r2)
    Sr1dot=np.zeros_like(Sq2)
    Sq2dot=Sr1dot
    Sr2dot=Sr1dot
    
    
          
        #Partial Velocity Matrices
    V1=np.zeros_like(w1)
    V2=np.zeros_like(w1)
          
    V1[:3,:]=np.dot(Sr1,San)
    V2[:3,:]=np.dot(Sq2,San)
    V2[3:,:]=np.dot(Sr2,Sbn)
    
               
        #Derivative of Partial Velocity Matrices
    V1dot=np.zeros_like(V1)
    V2dot=np.zeros_like(V2)
          
    V1dot[:3,:]=np.dot(Sr1,Sandot)+np.dot(Sr1dot,San)
    V2dot[:3,:]=np.dot(Sq2,Sandot)+np.dot(Sq2dot,San)
    V2dot[3:,:]=np.dot(Sr2,Sbndot)+np.dot(Sr2dot,Sbn)
    
          
        # w transformation matrix
    w=np.zeros((3*N,3*N))
    w[:3,:3]=San
    w[:3,3:]=San
    w[3:,3:]=Sbn
    
         
        #y coordinates
    y=np.dot(xgendot,w)
           
        #Equations of Motion
        #no Inertia matrices in a because point masses
    a=m1*np.dot(np.dot(w,V1),np.transpose(np.dot(w,V1)))+m2*np.dot(np.dot(w,V2),np.transpose(np.dot(w,V2)))
          
        #again no Inertial matrices
    b=m1*np.dot(np.dot(w,V1),np.transpose(np.dot(w,V1dot)))+m2*np.dot(np.dot(w,V2),np.transpose(np.dot(w,V2dot)))
            
        #c is zero from no inertias
            
    f1=np.array((0,-1*m1*9.81,0))
    f2=np.array((0,-1*m2*9.81,0))
          
    f=np.dot(np.dot(w,V1),f1)+np.dot(np.dot(w,V2),f2)
        
            
        #M is a matrix that I use to create equations to integrate
            
    M=np.zeros((n,5))
    M[0,0]=a[2,2]
    M[0,1]=a[2,5]
    M[1,0]=a[5,2]
    M[1,1]=a[5,5]
    M[0,2]=b[2,2]
    M[0,3]=b[2,5]
    M[1,2]=b[5,2]
    M[1,3]=b[5,5]
    M[0,4]=f[2]
    M[1,4]=f[5]
    M[0,:]=M[0,:]/M[0,0]
    M[1,:]=M[1,:]-(M[0,:]*M[1,0])
    M[1,:]=M[1,:]/M[1,1]
    M[0,:]=M[0,:]-(M[1,:]*M[0,1])
           
    d_dt=np.zeros((4))
    d_dt[0]=xdot[0]
    d_dt[1]=xdot[1]
    d_dt[2]=M[0,4]-xdot[0]*M[0,2]-xdot[1]*M[0,3]
    d_dt[3]=M[1,4]-xdot[0]*M[1,2]-xdot[1]*M[1,3]
    return d_dt
    
    
            
    






```

```python
#Integrate function, check for correctness, and plot results

yy=odeint(funct,xinit,Time)
theta1=yy[:,0]
theta2=yy[:,1]+theta1
theta1dot=yy[:,2]
theta2dot=yy[:,3]+theta1dot
V=-(m1+m2)*g*la*np.cos(theta1)-m2*g*lb*np.cos(theta2)
T=.5*m1*(la**2)*(theta1dot**2)+.5*m2*((la**2)*(theta1dot**2)+(lb**2)*(theta2dot**2)+2*la*lb*theta1dot*theta2dot*np.cos(theta1-theta2))
E=V+T
print("Difference in energy:")
print((max(E)-min(E)))
plt.plot(Time,V+T)
plt.title("Energy of System")
plt.xlabel("Time [s]")
plt.ylabel("Energy[J]")
plt.show()
plt.plot(Time,theta1,Time,theta2)
plt.xlabel("Time [s]")
plt.ylabel("Angle [Rad]")
plt.title("System Response")
plt.legend(("Theta1","Theta2"),"best")

plt.show()

```
