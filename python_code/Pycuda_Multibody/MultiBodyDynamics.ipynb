```python
#This script solves for the accelleration and velocity of a point when it's velocity and accelleration are known at an intermediate
#reference frame.  At the moment this script only deals with systems that do not have any branches, and pycuda has not yet been
#implemented.  This code isn't done yet and will continue to be improved.

import numpy as np
import MultiBodyFuncts as MBF

timesteps=200
numRefs=3
Time=np.linspace(0,50,timesteps)
theta=np.zeros((timesteps))
thetadot=np.zeros_like(theta)
theta=3.14*Time
thetadot+=3.14

l1=3
l2=(1+np.sin(theta))
l2dot=np.cos(theta)*thetadot
l2dubdot=(-1*np.sin(theta)*(thetadot**2))
S=np.zeros((numRefs,numRefs,3,3,timesteps))
a1=theta-(3.14/2)
b1=np.zeros_like(a1)
g1=np.zeros_like(a1)
a2=(3.14/2)-theta
b2=np.zeros_like(a1)
g2=np.zeros_like(a1)
angles=np.zeros((3,numRefs-1,timesteps))
angles[:,0,:]=np.array((a1,b1,g1))
angles[:,1,:]=np.array((a2,b2,g2))
x=0
#This part gets the S matrices
for t in range(0,timesteps):
    y=1
    for x in range(0,numRefs-1):
            a=angles[:,x,t]
            S[y,x,:,:,t]=MBF.directionCosine(a[0],a[1],a[2])
          
            y+=1

for t in range(0,timesteps):
    for x in range(2,numRefs):
        S[x,0,:,:,t]=np.dot(S[x,x-1,:,:,t],S[x-1,0,:,:,t])
for t in range(0,timesteps):
    for x in range(0,numRefs):
        for y in range(0,numRefs):
            S[x,y,:,:,t]=np.transpose(S[y,x,:,:,t])
            
#this part is where the important distances, angular velocities, and velocities are set#########################
r=np.zeros((numRefs,1,3,timesteps))#r holds the distance from refx to refx+1 in the coordinates of refx
                                    #this part requires the coder to define the system
for t in range(0,timesteps):
    r[0,0,:,t]=np.dot(S[0,1,:,:,t],np.array((0,l1,0)))
    r[1,0,:,t]=np.array((0,0,0))
    r[2,0,:,t]=np.array((0,0,-l2[t]))

V=np.zeros_like(r)#This matrix gives the velocity of point A with respect to refx
for t in range(0,timesteps):#coder just fills the last vector at every time t
    V[numRefs-1,0,:,t]=np.array((0,0,-l2dot[t]))
    
    
w=np.zeros((numRefs-1,1,3,timesteps))#w hold the angular velocities of the reference frames with respect to the one before it
wdot=np.zeros_like(w)#the angular accelleration
for t in range(0,timesteps):#there is one less row of vectors because the last reference frame doesn't have another one to point to
    w[0,0,:,t]=np.array((thetadot[t],0,0))
    w[1,0,:,t]=np.array((-thetadot[t],0,0))

    

########################################################################################################################
#    Now we put things in different reference frames.  All probably aren't needed but then I can see what will work best    
    
R=np.zeros_like(r)#R is the same as r, but all distances are given in the inertial reference frame
R[0,0,:,:]=r[0,0,:,:]

for t in range(0,timesteps):
    for x in range(1,numRefs):
        R[x,0,:,t]=np.dot(S[0,x,:,:,t],r[x,0,:,t])

RA=np.zeros_like(r)#Ra is the distance from refx to point A with respect to refx
RAn=np.zeros_like(r)#same as RA but all vectors given in newtonian frame
RA[numRefs-1,0,:,:]=r[numRefs-1,0,:,:]
RAn[numRefs-1,0,:,:]=R[numRefs-1,0,:,:]
for t in range(0,timesteps):
    for x in range(1,numRefs):
        RA[numRefs-1-x,0,:,t]=np.dot(S[numRefs-1-x,numRefs-x,:,:,t],RA[numRefs-x,0,:,t])+r[numRefs-1-x,0,:,t]
        RAn[numRefs-1-x,0,:,t]=RAn[numRefs-x,0,:,t]+R[numRefs-1-x,0,:,t]
wn=np.zeros_like(w)#will hold w but everything is in n
wn[0,0,:,:]=w[0,0,:,:]
for t in range (0,timesteps):
    for x in range(1,numRefs-1):
        wn[x,0,:,t]=np.dot(S[0,x,:,:,t],w[x,0,:,t])
Vn=np.zeros_like(V)
for t in range(0,timesteps):
    Vn[numRefs-1,0,:,t]=np.dot(S[0,numRefs-1,:,:,t],V[numRefs-1,0,:,t])

##########################################################################################################################
#This is where the velocity and Accellerations are found.  The velocity is found in two ways, mainly to ensure that 
#Everything is working correctly.
for t in range(0,timesteps):
    for x in range(1,numRefs):
        V[numRefs-1-x,0,:,t]=np.dot(S[numRefs-1-x,numRefs-x,:,:,t],V[numRefs-x,0,:,t])+np.cross(w[numRefs-1-x,0,:,t],RA[numRefs-1-x,0,:,t])
    
for t in range(0,timesteps):
    for x in range(1,numRefs):
        Vn[numRefs-1-x,0,:,t]=Vn[numRefs-x,0,:,t]+np.cross(wn[numRefs-1-x,0,:,t],RAn[numRefs-1-x,0,:,t])
A=np.zeros_like(Vn)
for t in range(0,timesteps):
    A[numRefs-1,0,:,t]=np.dot(S[0,numRefs-1,:,:,t],np.array((0,0,-l2dubdot[t])))

for t in range(0,timesteps):
    for x in range(1,numRefs):
        A[numRefs-1-x,0,:,t]=A[numRefs-x,0,:,t]+(2*np.cross(wn[numRefs-1-x,0,:,t],Vn[numRefs-x,0,:,t]))+np.cross(wn[numRefs-1-x,0,:,t],np.cross(wn[numRefs-1-x,0,:,t],RAn[numRefs-1-x,0,:,t]))+np.cross(wdot[numRefs-1-x,0,:,t],RAn[numRefs-1-x,0,:,t])
```
